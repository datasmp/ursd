# Close Data

```python
STOCK_DICT = {'Apple': 'AAPL', 'Tesla': 'TSLA', 'Amazon': 'AMZN', 'Visa': 'V', 'Microsoft': 'MSFT'}
START = '2015-1-1'
END = '2020-12-31'
```

```python
df_close = pd.DataFrame()

for name, symbol in STOCK_DICT.items():
    df_close[name] = yf.Ticker(symbol).history(start=START, end=END).Close
    
df_close.head().round(2)
```

```python
df_close.to_csv('close_stock_data_raw.csv')
```
